##### Move Zeroes
* **Description** : A chocolate factory is packing chocolates into packets. The chocolate packets here represent an array of N integer values. The task is to find the empty packets (0) of chocolate and push them to the end of the conveyor belt (array).
* **Input Format**: First line contains an integer n, representing size of an array. Second line contains n space separated integer values.
* **Output Format**: The output is the array with all zeros moved to the end, while maintaining the order of non-zero elements.
* **Sample Case 1**
    * Input:
    ```text
        8
        4,5,0,1,9,0,5,0
    ```
    * Output
    ```text
        4,5,1,9,5,0,0,0
    ```




````python
def zero_pos(arr):

    last_non_zero = 0
    for i in range(len(arr)-1):
        if arr[i] != 0:
            # Swap current element with the 'last_non_zero' position
            arr[last_non_zero], arr[i] = arr[i], arr[last_non_zero]
            last_non_zero += 1

    return  arr

# size = int(input("Enter Array size"))
# elem_list = [int(input(f"Enter element {i+1} -> ")) for i in range(size)]

arr = zero_pos ([4,5,0,1,9,0,5,0])
print(arr)
````
